```python
data = {
    'Energy Sources':['solar','wind','hydropower','geothermal','biomass','tidal'],
    'Energy Consumption(MW)':[1200,np.nan,2900,np.nan,2500,3200],
    'Cost (million $)':[200,400,np.nan,150,250,np.nan]
}
```

```python
e_df = pd.DataFrame(data)
e_df.head()
```

```python
e_df = pd.DataFrame(data)
e_df
```

```python
#missing value - isnull()
e_df.isnull()       #Not Good
```

```python
e_df.isnull().sum()
```

```python
#Handling missing value - imput drop
#dropna
df_cleaned = e_df.dropna()
df_cleaned
```

```python
e_df ['Energy Consumption(MW)']=e_df['Energy Consumption(MW)'].fillna(e_df['Energy Consumption(MW)'].mean())
e_df ['Cost (million $)']=e_df['Cost (million $)'].fillna(e_df['Cost (million $)'].mean())
e_df
```

```python
 #Normalization -minmax scaler
 from sklearn.preprocessing import MinMaxScaler
 scaler = MinMaxScaler()
 e_df[['Energy Consumption(MW)','Cost (million $)']] = scaler.fit_transform(e_df[['Energy Consumption(MW)','Cost (million $)']])
 e_df
```

```python
#Z Score-Standard - mean=0, std=1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
e_df[['Energy Consumption(MW)','Cost (million $)']] = scaler.fit_transform(e_df[['Energy Consumption(MW)','Cost (million $)']])
e_df
```

```python
#Encoding - categorical variable to numerical variable
e_encoded = pd.get_dummies(e_df,columns=['Energy Sources'])
e_encoded
```

```python
#Feature Engineering:New feature = energy cons per million $
e_encoded['Consumption per million $'] = e_df['Energy Consumption(MW)']/e_df['Cost (million $)']
e_encoded
```

Visualization

```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
en_cons = [1200, 1300, 1100, 1500, 2600, 1700]

plt.plot(months,en_cons,color='blue',marker='o',linestyle='dashed')
plt.title('Plot EC vs Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
#Bar Plot
en_sources = ['solar','wind','hydropower','geothermal','biomass']
en_values = [1200,3400,2900,1800,2500]

plt.bar(en_sources,en_values,color='orange')
plt.title('Plot EC vs Energy Sources')
plt.xlabel('Energy Sources')
plt.ylabel('Energy Consumption (MW)')
plt.show()
```

```python
#Pie Chart
plt.pie(en_values, labels=en_sources, autopct="1.2f%%")
plt.title('Energy Consumption by sources')
plt.show()
```

```python
#Scatter Plot
carbon_emission = [200, 500, 450, 300,250]

plt.scatter(en_values, carbon_emission, color='blue')
plt.title('Energy Consumption vs Carbon Emission')
plt.xlabel('Energy Consumption (MW)')
plt.ylabel('Carbon Emission (tons)')
plt.show()
```

```python
plt.bar(en_sources, en_values, color='red', edgecolor='black',linewidth=2)
plt.grid(True,linestyle='--',alpha=0.7)
plt.legend(['Energy Consumption'],loc='upper left')
plt.title('Customized plot',fontsize=16,fontweight='bold')
plt.xlabel('Energy Sources',fontsize=14)
plt.ylabel('Energy Consumption (MW)',fontsize=14)
plt.show()
```
